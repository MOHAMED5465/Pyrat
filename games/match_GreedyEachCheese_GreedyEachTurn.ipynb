In this script, we will compare `GreedyEachTurn`,`GreedyEachCheese` in a 100 games . \
The goal is to see wich player has the best performance by playing a 100 games.

```python
# External imports
import sys
import os
import tqdm.auto as tqdm
import matplotlib.pyplot as pyplot
import scipy.stats as scstats

# Add needed directories to the path
sys.path.append(os.path.join("..", "players"))

# PyRat imports
from   pyrat import *
from GreedyEachTurn_density import GreedyEachTurn_density
from two_opt_reactive import Two_opt_reactive 
import statistics
```

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">CONSTANTS</h1>

In this script, we are going to make 100 independent matches. \
We will calculate the number of wins for each player,in order to compare their \
performances .

```python
# Determines how many games will be played for each player
NB_GAMES = 100
```

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">RUN THE GAMES</h1>

Let us now perform all games. \
For each game, we will store the winner.

```python
# Players to test (keys are legends to appear in the plot)
players = {"GreedyEachTurn_density": {"class":GreedyEachTurn_density , "args": {}},
           "Two_opt_reactive": {"class": Two_opt_reactive , "args": {}},
            }
results={}
for i in range(NB_GAMES):
    results[i]={"GreedyEachTurn_density":None,"Two_opt_reactive":None}
#create 100 games between palyers with different random seed     
for seed in tqdm.tqdm(range(NB_GAMES)):

    CONFIG = {'maze_width': 25,
          'maze_height': 20,
          'cell_percentage': 80.0,
          'wall_percentage': 60.0,
          'mud_percentage': 20.0,
          'mud_range': [4, 9],
          'preprocessing_time': 3.0,
          'turn_time': 0.1,
          'random_seed':seed,
          'render_mode':RenderMode.NO_RENDERING}
    # Instantiate a game with specified arguments
    game = Game(**CONFIG)
    # Instantiate players in distinct teams
    player_1=GreedyEachTurn_density()
    player_2=Two_opt_reactive()
    game.add_player(player_1, team="Team Pythonz", location=StartingLocation.RANDOM)
    game.add_player(player_2, team="Team Ghostz", location=StartingLocation.RANDOM)
    # Start the game
    stats = game.start()
    for key in players:
        results[seed][key]=stats['players'][key]['score']

    print(results)

#calculate the number of wins for each player
nb_wins_GreedyEachTurn_density=0
nb_wins_Two_opt_reactive=0

for key in results:
    L=list(results[key].items())
    print(L)
    max_score=max(L, key=lambda x: x[1])[1]
    max_players = [t for t in L if t[1] == max_score]
    if 'GreedyEachTurn_density'in max_players and 'Two_opt_reactive' in max_players:
        nb_wins_GreedyEachTurn_density+=0.5
        nb_wins_Two_opt_reactive+=0.5
    elif 'GreedyEachTurn_density'in max_players :
        nb_wins_GreedyEachTurn_density+=1
    elif 'Two_opt_reactive' in max_players:
        nb_wins_Two_opt_reactive+=1

#print results 
print
print(nb_wins_GreedyEachTurn_density)
print(nb_wins_Two_opt_reactive)
```
